```python
import numpy as np
import matplotlib.pyplot as plt
x_data = [1.0, 1.8, 3.0, 4.1, 5.2, 6.0]
y_data = [1.0, 1.3, 2.2, 2.5, 2.8, 3.6]
x1_data = [1.0, 6.0]
y1_data = [1.0, 3.6]
```

```python
x = np.array(x_data)
y = np.array(y_data)
x1 = np.array(x1_data)
y1 = np.array(y1_data)
plt.scatter(x, y)
plt.scatter(x1, y1)
plt.plot(x, y, 'r')
plt.plot(x1, y1, 'g')
plt.grid()
plt.show()
```
