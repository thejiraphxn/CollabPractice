```python
class _PriorityQEntry(object):
    def __init__(self, item, priority):
        self.item = item
        self.priority = priority

class UnPriorityQueue:
    def __init__(self):
        self.qList = []
        self.size = 0

    def isEmpty(self):
        return len(self.qList) == 0

    def __len__(self):
        return len(self.qList)

    def enqueue(self, item, priority):
        entry = _PriorityQEntry(item, priority)
        self.qList.append(entry)
        self.size += 1

    def dequeue(self):
        assert not self.isEmpty(), "Cannot dequeue from an empty queue."
        highest_prio_index = 0
        for i in range(1, len(self.qList)):
            if self.qList[i].priority < self.qList[highest_prio_index].priority:
                highest_prio_index = i
        entry = self.qList.pop(highest_prio_index)
        return entry.item

Qlist = UnPriorityQueue()
print("Check Qlist is empty or not : ",Qlist.isEmpty())

DataQueue = [["Green", 6], ["Black", 7], ["Cyan", 3], ["White", 0], ["Blue", 2], ["Red", 9], ["Maroon", 4]]
for i in range (len(DataQueue)):
  Qlist.enqueue(DataQueue[i][0], DataQueue[i][1])

print("Qlist is empty or not after enqueue : ", Qlist.isEmpty())
print("Size of Qlist after enqueue : ", Qlist.__len__())
print("Data of dequeue")

while not Qlist.isEmpty():
    print(Qlist.dequeue())

print("Qlist is empty or not after dequeue : ", Qlist.isEmpty())
print("Size of Qlist after dequeue : ", Qlist.__len__())

Qlist.dequeue()
```
