```python
def hr(header):
  if(header):
      print(f'------------------ {header} ------------------------')
  else:
      print('--------------------------------------------------')


hr('Start Tuple')
Tuple = ('abc', 'def')
print(Tuple)

hr(False)

#creating a tuple with repeatition
tp = ('CPE499',) * 5
print(tp)

hr(False)

#Accessing tuple with index
print(Tuple[0])

hr(False)

tpString = tuple('PYTHON')
print(tpString)
print(tpString[1:4])
print(tpString[:4])

hr(False)

#Tuple unpacking
tpUnpack = ('Python', 'C#', 'C/C++')
a, b, c = tpUnpack
print(a, b, c)

hr(False)

#Concatenation of Tuple
tp1 = (1, 2, 3)
tp2 = ('A', 'B', 'C')
tp3 = tp1 + tp2
print(tp3)

hr(False)

#Slicing of Tuple
tpSlicing = tuple('Python')
print(tpSlicing)
print(tpSlicing[1:4])

hr(False)

#Delete Tuple
DelTuple = (1, 2, 3, 4, 5, 6)
print(DelTuple)
del DelTuple
#It will return error print(DelTuple)

hr(False)

```

```python
from typing import Union
from re import U
#Start set
Set = {1, 2, 3, 4}
print(Set)


hr('Using set function')
#Using set function
FuncSet = set()
print(FuncSet)

FuncSet = set('ILovePython') #String to set
print(FuncSet)

FuncSet = set(['I', 'hate', 'python']) #List to set
print(FuncSet)

FuncSet = ('I', 'hate', 'python') #Tuple to se4t
print(set(FuncSet))

FuncSet = {'id': 1, 'number': 3} #Dict to set
print(set(FuncSet))

hr('Add elements to set')

#Add elements to set

EleSet = {1, 2, 3, 4, 5, 6}
print(EleSet)
EleSet.add(7)
print(EleSet)
EleSet.update([8, 9, 10])
print(EleSet)

hr('Pop method')

#Pop methos

PopSet = {1, 2, 3, 4, 5, 6}
print(PopSet)
PopSet.pop()
print(PopSet)

hr('Clear method')

#Clear method
ClearSet = {1, 2, 3, 4, 5, 6}
print(ClearSet)
ClearSet.clear()
print(ClearSet)

hr('Remove method')

#Remove method
RemSet = {1, 2, 3, 4, 5, 6}
print(RemSet)
RemSet.remove(3)
print(RemSet)

hr('Discard method')

#Discard method
DisSet = {1, 2, 3, 4, 5, 6}
print(DisSet)
DisSet.discard(3)
print(DisSet)


#Union method
hr('Union method')
UnionSet = {1, 2, 3, 4, 5, 6}
print(UnionSet)
UnionSet2 = {7, 8, 9, 10}
print(UnionSet2)
print(UnionSet.union(UnionSet2))
```

```python
#Start Dictionary
hr('Dictionary')
Dict = {'id': 1, 'firstname': 'John', 'lastname': 'Doe', 'skill': ['python', 'javascript', 'mongodb']}
print(Dict['skill'])


for skill in Dict['skill']:
  print(skill)

hr('Accessing Dictionary')
AccessDict = {'id': 1, 'firstname': 'John', 'lastname': 'Doe', 'skill': ['python', 'javascript', 'mongodb']}
print(AccessDict['firstname'])
print(AccessDict.get('lastname'))
#in dict 0, 1 if True or False existed it will select True or Flase to result

hr('Update Dictionary')
UpdateDict = {'id': 1, 'firstname': 'John', 'lastname': 'Doe', 'skill': ['python', 'javascript', 'mongodb']}
UpdateDict['firstname'] = 'Jane'
UpdateDict['lastname'] = 'Roe'
print(UpdateDict)
UpdateDict['skill'].append('C++')
UpdateDict['skill'].append('C#')
UpdateDict['skill'].append('Java')

hr('Remove Values')
#del, pop(), clear(), popitem()
UpdateDict['skill'].remove('python')
del UpdateDict['skill']
print(UpdateDict)
UpdateDict.pop('id')
print(UpdateDict)
DKey, DVal = UpdateDict.popitem()
print(f'Key: {DKey} | Val: {DVal}')
UpdateDict.clear()
print(UpdateDict)


#Nest Dictionary
hr('Nest Dictionary')
NestDict = {1: 'Python', 2:'Javascript', 3:'C#', 4:'C++', 5:{'A': 'Microsoft', 'B': 'Apple', 'C': 'Samsung'}}
print(NestDict[5]['A'])
NestFruit = {'Fruit': {'Apple': {'Color': 'Red'}, 'Orange': {'Color': 'Orange'}}}
print(NestFruit['Fruit']['Apple'])
print(NestDict[3])
```

```python
#Arithmetic Operators
hr('Arithmetic Operators')
a = 100
b = 200
print(f'addition a + b = {a + b}')
print(f'subtraction a - b = {a - b}')
print(f'multiplication a * b = {a * b}')
print(f'division a / b = {a / b}')
print(f'modulus a % b = {a % b}')
print(f'exponent a ** b = {a ** b}')
print(f'floor division a // b = {a // b}')

#Comparison
hr('Comparison')
print(f'comparison a > b = {a > b}')
print(f'comparison a < b = {a < b}')
print(f'comparison a == b = {a == b}')
print(f'comparison a != b = {a != b}')
print(f'comparison a >= b = {a >= b}')
print(f'comparison a <= b = {a <= b}')

#Bitwise
hr('Bitwise')
print(f'bitwise a & b = {a & b}')
print(f'bitwise a | b = {a | b}')
print(f'bitwise a ^ b = {a ^ b}')
print(f'bitwise ~a = {~a}')
print(f'bitwise a << b = {a << b}')
print(f'bitwise a >> b = {a >> b}')


#Identity
hr('Identity')
print(f'identity a is b = {a is b}')
print(f'identity a is not b = {a is not b}')

#Membership
hr('Membership')
i = 5
msList = [1, 2, 3, 4, 5, 6, 7, 8, 9]
if(i in msList):
  print(f'{i} is in list')
else:
  print(f'{i} is not in list')


#Ternary
#Syntax [on_true] if [expression] else [on_false]
a, b = 10, 20
min = a if a < b else b # min is a when a < b then false min will be b
print(min)

#Operator Associativity in Python
hr('Operator Associativity')
print(f'{100/10*10}')
print(f'{5 - 2 + 3}')
print(f'{5-(2 + 3)}') #in braces is first piority
print(f'{(5 - 2) + 3}') #in braces is first piority
print(f'{2 ** 3 ** 2}')
```
