```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/CPE270/playTennis.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
def LabEn(data, column):
  for i in column:
    lb = LabelEncoder().fit_transform(data[i])
    data[i] = lb

columns = ['outlook', 'temp', 'humidity', 'windy', 'play']
LabEn(df, columns)
print(df)

x = df[['outlook', 'temp', 'humidity', 'windy']]
y = df['play']
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x, y)
class_name = ['No', 'Yes']
data_new = [[1, 2, 0, 0]]
y_predict = model.predict(data_new)
y_new = class_name[y_predict[0]]
print(y_new)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize = (25, 20))
t = plot_tree(model, feature_names=x.columns, class_names=class_name, impurity=True, precision=3, filled=True, rounded=True, fontsize=16)
plt.show()
```
