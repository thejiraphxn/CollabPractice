```python
class Employee:
    def __init__(self, id, firstname, lastname, salary, department):
        self.firstname = firstname
        self.lastname = lastname
        self.id = id
        # self.__salary = salary
        self.__salary = salary #__ Double underscore is private varriable, prop
        self.department = department


    def emp(self):
      if self.department == 'IT':
        print('Department: Information Technology')
      elif self.department == 'HR':
        print('Department: Human Resource')
      else:
        print('Department: ', self.department)

    def calculate_salary(self, salary, hour_worked):
      overtime = 0
      if hour_worked > 50:
        overtime = hour_worked - 50
        self.salary = self.salary + (overtime * (self.salary / 30))

    def assign_department(self, emp_dept):
      self.department = emp_dept

    def print_employee_details(self):
      print("\nName: ", self.firstname, self.lastname)
      print("Employee ID: ", self.id)
      print("Salary: ", self.salary)
      print("Department: ", self.department)

    def __displaySalary(self):
      print(self.salary)

    def setSalary(self, salary):
      self.salary = salary

    def getSalary(self):
      return self.salary


emplP1 = Employee("001", "John", "Doe", 45000, "IT")
emplP2 = Employee("002", "Ryan", "Doe", 55000, "HR")
emplP1.emp()
emplP2.emp()
emplP1.setSalary(60000)
emplP2.setSalary(60000)
print("P1:", emplP1.id, emplP1.firstname, emplP1.lastname, emplP1.getSalary(), emplP1.department)
print("P2:", emplP2.id, emplP2.firstname, emplP2.lastname, emplP2.getSalary(), emplP2.department)

emplP1.calculate_salary(45000, 52)
emplP2.calculate_salary(55000, 60)
emplP1.assign_department("Developer")
emplP2.assign_department("Finance")
print("P1:", emplP1.id, emplP1.firstname, emplP1.lastname, emplP1.salary, emplP1.department)
print("P2:", emplP2.id, emplP2.firstname, emplP2.lastname, emplP2.salary, emplP2.department)
```

```python
class Computer:
  def __init__(self, cpu='', memory=0, local_storage=0):
    self.__cpu = cpu
    self.__memory = memory
    self.__local_storage = local_storage

  def getCPU(self):
    return self.__cpu

  def setCPU(self, cpu):
    self.__cpu = cpu

  def getMemory(self):
    return self.__memory

  def setMemory(self, memory):
    self.__memory = memory

  def getLocal_storage(self):
    return self.__local_storage

  def setLocal_storage(self, local_storage):
    self.__local_storage = local_storage

Com1 = Computer()
Com1.setCPU('Intel Core I7')
Com1.setMemory(32)
Com1.setLocal_storage(1000)
print(f'CPU: {Com1.getCPU()}')
print(f'Ram: {Com1.getMemory()} GB')
print(f'Storage: {Com1.getLocal_storage()} GB')
```

```python
#Write as @property no get set

class CellPhone:
    def __init__(self, cpu, brand, model, memory, storage):
        self.__cpu = cpu
        self.__brand = brand
        self.__model = model
        self.__memory = memory  # Using setter for validation
        self.__storage = storage  # Using setter for validation

    @property
    def brand(self):
        return self.__brand

    @brand.setter
    def brand(self, brand):
        self.__brand = brand

    @property
    def cpu(self):
        return self.__cpu

    @cpu.setter
    def cpu(self, cpu):
        self.__cpu = cpu

    @property
    def model(self):
        return self.__model

    @model.setter
    def model(self, model):
        self.__model = model

    @property
    def memory(self):
        return self.__memory

    @memory.setter
    def memory(self, memory):
        self.__memory = memory

    @property
    def storage(self):
        return self.__storage

    @storage.setter
    def storage(self, storage):
        self.__storage = storage


# Create an instance of CellPhone with proper arguments
cell_phone = CellPhone('Apple A15', 'Apple', 'iPhone 14', 8, 256)

# Print the values
print(cell_phone.brand)    # Apple
print(cell_phone.cpu)      # Apple A15
print(cell_phone.model)    # iPhone 14
print(cell_phone.memory)   # 8
print(cell_phone.storage)  # 256
```

```python
# static method
class Circle:
  PI = 22/7

  @staticmethod
  def area(radius):
    return Circle.PI * radius * radius

  @staticmethod
  def perimeter(radius):
    return 2 * Circle.PI * radius

  @staticmethod
  def diameter(radius):
    return 2 * radius

  @staticmethod
  def circumference(radius):
    return 2 * Circle.PI * radius

print(Circle.area(5))
print(Circle.perimeter(5))
print(Circle.diameter(5))
print(Circle.circumference(5))
```

```python
# class method
import math
class Circle:
  PI = 22/7
  def __init__(self, radius):
    self.radius = radius

  @classmethod
  def radius(self):
    return self.radius

  @classmethod
  def from_area(cls, area):
    return cls(math.sqrt(area/Circle.PI))

  @classmethod
  def from_circumference(cls, circumference):
    return cls(circumference/(2*Circle.PI))

  @classmethod
  def from_diameter(cls, diameter):
    return cls(diameter/2)


circle1 = Circle.from_area(786.85775)
circle2 = Circle.from_circumference(100)
circle3 = Circle.from_diameter(10)
print(circle1.radius)
print(circle2.radius)
print(circle3.radius)
```

```python
# inheritance

class ParentClass:
  parentMsg = 'This is parent class !'
  def __init__(self):
    self.Name = 'John'
    self.Age = 25

class ChildClass(ParentClass):
  childMsg = 'This is child message !'

child = ChildClass()
print(child.parentMsg)
print(child.childMsg)
print(child.Name)
print(child.Age)
```

```python
class GrandParentClass:
  def __init__(self, a):
    print('init GrandParentClass')
    self.a = a
    print("a = ", self.a)

class ParentClass(GrandParentClass):
  def __init__(self, a, b):
    print('init ParentClass')
    super().__init__(a) # ดึงค่าจาก Grand
    print('init ParentClass')
    self.b = b
    print("b = ", self.b)

class ChildClass(ParentClass):
  def __init__(self, a, b, c):
    print('init ChildClass')
    super().__init__(a, b)
    print('init Child')
    self.c = c
    print("c = ", self.c)

child = ChildClass(1, 2, 3)
```
