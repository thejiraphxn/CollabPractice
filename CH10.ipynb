```python
import numpy as np
input = [1, 2, 3, 4]
weights = [0.3, 0.4, 0.8, 0.5]
y1 = np.dot(input, weights)
y = 1/(1+math.exp(-y1))
print(y)
yt = 1
err = (math.pow((y-yt), 2)/2)
print("Error = ", "{:.10f}".format(err))
```

```python
import math
import numpy as np
input = [1, 3, 4, 2]
weights = [(1, 0.2, 0.7, 0.2), (1, 0.5, 0.8, 0.7), (0.4, 0.2)]
y1 = np.dot(input, weights[0])
y2 = np.dot(input, weights[1])
y1_new = 1/(1+math.exp(-y1))
y2_new = 1/(1+math.exp(-y2))
y3 = np.dot((y1_new, y2_new), weights[2])
y = 1/(1+math.exp(-y3))
print(y1_new, y2_new, y)
yt = 0.5
err = (math.pow((y-yt), 2)/2)
print("Error = ", "{:.10f}".format(err))
```

```python
from sklearn.linear_model import Perceptron
import pandas as pd
path = "/content/drive/MyDrive/CPE270/perceptron.csv"
df = pd.read_csv(path)
x = df[['x1', 'x2']]
y = df['y']
model = Perceptron()
model.fit(x, y)
x_new = [[1.0, 1.5]]
y_predict = model.predict(x_new)
print("y_new =", y_predict[0])
print("weights =", model.coef_)
print("bias =", model.intercept_)
```
