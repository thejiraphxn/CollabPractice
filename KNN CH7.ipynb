```python
from sklearn.metrics.pairwise import euclidean_distances
x = [[3,3], [5,8], [4,7], [6,4], [5,9], [7,5], [6,9], [4,5], [3,8], [2,6]]
y = ['A', 'B', 'B', 'B', 'A', 'B', 'B', 'A', 'B', 'A']
x_new = [[5,5]]
distance = euclidean_distances(x, x_new)
print('id       distance            class')
for (i, d) in enumerate(distance):
  print(i+1, '\t', '{:2f}'.format(d[0]), '\t\t', y[i])
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/CPE270/shirt_size.xlsx")
x = df[['Height_cms', 'Weight_kgs']]
y = df['T_Shirt_Size']
k = 5
model = KNeighborsClassifier(n_neighbors= k)
model.fit(x, y)
x_new = [[161, 61]]
y_new = model.predict(x)
h = x_new[0][0]
w = x_new[0][1]
s = y_new[0]
print("K :", k)
print('Height_cms :', h, 'Weight_kgs :', w)
print("T-Shirt size :", s)
acc = model.score(x, y)*100
print("Accurate : {:.2f}".format(acc), "%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_excel("/content/drive/MyDrive/CPE270/shirt_size.xlsx")
x = df[['Height_cms', 'Weight_kgs']]

scaler = StandardScaler()
x = scaler.fit_transform(x)
y = df['T_Shirt_Size']
k = 5

model = KNeighborsClassifier(n_neighbors= k)
model.fit(x, y)
x_new = [[161, 61]]
x_new_sc = scaler.transform(x_new)
y_new = model.predict(x_new_sc)

h = x_new[0][0]
w = x_new[0][1]
s = y_new[0]

print("K :", k)
print('Height_cms :', h, '\nWeight_kgs :', w)
print("T-Shirt size :", s)
acc = model.score(x, y)*100
print("Accurate : {:.2f}".format(acc), "%")
```
