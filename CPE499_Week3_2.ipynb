```python
def hr(header):
  if(header):
      print(f'------------------ {header} ------------------------')
  else:
      print('--------------------------------------------------')

hr('Looping in Python can use for list tuples strings and range')
#Looping in Python can use for list tuples strings and range

hr('For-Loop')
#For-Loop
s = 'Python'
for i in s:
  print(i)

hr('For-Loop include range')
#Using range
hr('range 1 param it will start at 0 range(stop)')
for i in range(5):
  print(i)

hr('range 2 params range(start, stop)')
for i in range(2, 5):
  print(i)

hr('range 3 params range(start, stop, stepup)')
for i in range(0, 20, 2):
  print(i)

hr('Control Statement in loop')
#Control Statement
for i in "I love Python":
  if i == 'l' or i == 'P':
    continue
  print(i) #Never print l and P cause continue will pass the print function

hr('Break with for loop')
#Break with for loop
for i in "I love Python":
  if i == 'l':
    break
  print(i) #It will stop loop and does not print anymore

hr('Pass Statement for loop')
#Pass Statement for loop
for i in "I love Python":
  if i == 'l':
    pass
  print(i)

hr('Else in loop')
#in python else condition be able to use as false statement
for i in range(1, 4):
  print(i)
else: print('Out of range cause condition is false')

hr('Using Enumerate() in loop')
#Enumerate()
EnumerateList = ['One', 'Two', 'Three']
for i in enumerate(EnumerateList):
  print(i)

#include index
my_list = ['i', 'you', 'we', 'they']
for index, val in enumerate(my_list):
  print(f'{index}, {val}')

hr('Nested For loops (Loop in loop)')
#Nested For loops
for i in range(1, 4):
  for j in range(1, 4):
    for k in range(1, 4):
      print(f'i:{i}, j:{j}, k:{k}')

hr('Question using loop')
BasketToWash = ['shirt', 'sock', 'pants', 'towel']
BasketWashing = []
for i in BasketToWash:
  if i == 'pants':
    break
  else:
    print(f'Washing : {i}')

BasketWashing.append('socks')
print(f'Washing : {BasketWashing}')


hr('Impolement range')
#Implement range
for i in range(1, 8):
  d = 3 + (i - 1) * 0.5
  print(f'Day {i}: Run {d:.1f} miles')
```
