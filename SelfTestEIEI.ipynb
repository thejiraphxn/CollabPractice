```python
class Person:
  def __init__(self, age, gender):
    self.name = ''
    self.age = age
    self.gender = gender

  def introduce(self):
    print(f"Hello, my name is {self.name}, {self.age}, {self.gender}")

```

```python
class Employee(Person):
  def __init__(self, jobtitle, salary):
    self.jobtitle = jobtitle
    self.salary = salary

  def work(self):
    print(f"Hello, my name is {self.name}, {self.jobtitle}, {self.salary}")

  def setName(self, name):
    self.name = name



P1 = Person(21, "Male")


P2.introduce()
```
