```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

path = "/content/drive/MyDrive/CPE270/hours_study.csv"
df = pd.read_csv(path)


x = df['hours_study'].values.reshape(-1, 1)
y = df['pass']

x_new = [2.2, 3.3, 4.4]
x_new = np.array(x_new).reshape(-1, 1)

model.fit(x, y)
y_predict = model.predict(x_new)

count = 0
for i in y_predict:
  print(f"input {x_new[count]} ", i)
  count = count + 1
```
