```python
def hr(header):
  if(header):
      print(f'------------------ {header} ------------------------')
  else:
      print('--------------------------------------------------')


hr('if else statement')
#Normal If else
age = 20
if age <= 20:
  print('Eligible to vote.')
else:
  print('You are not eligible to travel.')


hr('Short hand if')
#Short hand if
age = 19
if age > 18: print('Eligible to vote.')


hr('If else Statement')
#If else Statement

age = 10
if age <= 12:
  print('Accept to Travel.')
else:
  print('You are not eligible to travel.')

hr('Short hand if else')
#Short hand if else
if age <= 12: print('Accept to Travel.')
else: print('You are not eligible to travel.')


hr('Single line if else statement')
#Single line if else statement
res = 'Accept to Travel.' if age <= 12 else 'You are not eligible to travel.'
print(res)

hr('if else-if else (elif) for multiple conditions')
#if else-if else (elif) for multiple conditions
# age = int(input('Type your age : '))
age = 10

if age <= 12:
  print('Child')
elif age <= 19:
  print('Teenager')
elif age <= 35:
  print('Teenager')
else:
  print('Adult')


hr('Nested if else')
#Nested if else
age = 70
is_member = True
if age >= 60:
  if is_member:
    print('30% Discount for member')
  else:
    print('20% Discount for normal user')
else:
  print('Discount unavailable for Adult !')


hr('Match-case statement Switch-case')
#Match-case statement Switch-case
number = 4
match number:
  case 1:
    print('One')
  case 2:
    print('Two')
  case 3 | 5:
    print('Three or Five')
  case _:
    print('Invalid Number')


```
