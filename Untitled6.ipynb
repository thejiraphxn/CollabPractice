```python
import numpy as np
# data = [[2.5, 1],
#         [2.5, 2],
#         [2.5, 3],
#         [2.5, 4],
#         [2.5, 5]]
data = [[]]

# print("data =", data)
data = np.array(data)
# print("data =", data)
# x2 = data[:, 0]
# print("data =", x2)
x = data[:, 0].reshape(-1, 1)
# print("data =", x)
y = data[:, 1].reshape(-1, 1)
print("data =", y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)

print("ค่า coefficient หรือค่า a = ", model.coef_[0])
print("ค่า intercept หรือค่า b = ", model.intercept_)
```

```python
from sklearn.metrics import r2_score

y_predicted = model.predict(x)

score = r2_score(y, y_predicted)

print("y_predicted = ", y_predicted)
print("R-Squared = ", score)
accuracy = "{:.2f}".format(score * 100)
print(f"ค่าความแม่นยำคือ {accuracy}%")
```
