```python
# Corrected method names: inorder_traversal(), preorder_traversal(), postorder_traversal(), level_order_traversal()
# Usage: result.append(node.data) instead of result.append(node)
# Usage: print(Nodes["A"].level_order_traversal()) instead of print(Nodes["A"].levelOrderTraversal())

from collections import deque

class TreeNode:
    """Represents a node in a tree."""

    def __init__(self, data):
        """Initialize the TreeNode with data and an empty list of children."""
        self.data = data
        self.children = []

    def add_child(self, child):
        """Add a child node to the current node."""
        self.children.append(child)

    def inorder_traversal(self):
        """Perform inorder traversal of the tree."""
        result = []
        if self.children:
            result += self.children[0].inorder_traversal()
        result.append(self.data)
        if len(self.children) > 1:
            for child in self.children[1:]:
                result += child.inorder_traversal()
        return result

    def preorder_traversal(self):
        """Perform preorder traversal of the tree."""
        result = [self.data]
        for child in self.children:
            result += child.preorder_traversal()
        return result

    def postorder_traversal(self):
        """Perform postorder traversal of the tree."""
        result = []
        for child in self.children:
            result += child.postorder_traversal()
        result.append(self.data)
        return result

    def level_order_traversal(self):
        """Perform level-order traversal (Breadth-first search) of the tree."""
        result = []
        queue = deque()
        queue.append(self)
        while queue:
            node = queue.popleft()
            result.append(node.data)
            for child in node.children:
                queue.append(child)
        return result

def print_tree(root, level=0):
    """Print the tree structure."""
    if root is None:
        return
    print("    " * level + str(root.data))
    for child in root.children:
        print_tree(child, level + 1)

def add_node(parent, data):
    """Add a new child node to the parent node."""
    new_node = TreeNode(data)
    parent.add_child(new_node)

def insert_node(parent, index, data):
    """Insert a new child node at the specified index of the parent node."""
    new_node = TreeNode(data)
    parent.children.insert(index, new_node)

def search(root, data, position=None):
    """Search for a node with the specified data in the tree."""
    if position is None:
        position = []
    if root is None:
        return None, None
    if root.data == data:
        return root, position
    for i, child in enumerate(root.children):
        result, child_position = search(child, data, position + [i])
        if result:
            return result, child_position
    return None, None

Nodes = {}
NodesData = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
for Node in NodesData:
  Nodes[Node] = TreeNode(Node)


Nodes["A"].add_child(Nodes["B"])
Nodes["A"].add_child(Nodes["C"])
Nodes["A"].add_child(Nodes["D"])
Nodes["B"].add_child(Nodes["E"])
Nodes["C"].add_child(Nodes["F"])
Nodes["C"].add_child(Nodes["G"])
Nodes["D"].add_child(Nodes["H"])
Nodes["F"].add_child(Nodes["I"])
Nodes["G"].add_child(Nodes["J"])

print("Tree Structure")
print_tree(Nodes["A"])

add_node(Nodes["J"], "M")

insert_node(Nodes["B"], 0, "K")

add_node(Nodes["I"], "N")

print("Tree Structure after adding and inserting nodes")
print_tree(Nodes["A"])

# Find Node N
result_node, position = search(Nodes["A"], "N")
if result_node is not None:
    print("This node position is :", position)
else:
    print("Node not found")

# Find Node X
result_node, position = search(Nodes["A"], "X")
if result_node is not None:
    print("This node position is :", position)
else:
    print("Node not found")

print()
print("Inorder Traversal : ")
print(Nodes["A"].inorder_traversal())
print("Preorder Traversal : ")
print(Nodes["A"].preorder_traversal())
print("Postorder Traversal : ")
print(Nodes["A"].postorder_traversal())
print("Level Order Traversal : ")
print(Nodes["A"].level_order_traversal())
```
