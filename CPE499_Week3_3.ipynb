```python
def hr(header = False):
  if(header):
      print(f'------------------ {header} ------------------------')
  else:
      print('--------------------------------------------------')

hr('def with param')
def alert(param = False):
  if param:
    print(f"!!! {param} !!!")
  else:
    print("No param")

alert('ALERT')

hr('def with param define type and return')
def add(num1: int, num2: int, num3: int) -> int:
  return num1 + num2 + num3

print(add(1, 2, 4))

hr('prime number')
def is_prime(n):
  if n in [2, 3]:
    return True
  if n % 2 == 0 or n == 1:
    return False
  r = 3
  while r * r <= n:
    if n % r == 0:
      return False
    r += 2
  return True

print(is_prime(11))
print(is_prime(10))

hr('function type of args')
#Default Arguments
def myFun(x, y=500):
  print(x, y)

myFun(x = 5)

hr('Positional Argument')
#Positional Argument
def nameAge(name, age):
  print("Hi, I'm ", name)
  print('My age is ', age)

print("case 1")
nameAge("Tay", 22)
print("case 2")
nameAge(22, "Chen")

hr('Arbitary Keyword Arg')
#Arbitary Keyword Arg
# *arg none-keyword
# **kargs keyword
def myArgs(*args):
  for arg in args:
    print(arg)

myArgs(1,2,3,4,5,6,7,8,9)

def myKArgs(**kwargs): #for dictionary
  for key, val in kwargs.items():
    print('%s == %s' % (key, val))

myKArgs(first = "A", mid = "B", last = "C")

hr('Nest Function ')
#Nest Function
def f1():
  s = "I Love Python"
  def f2():
    print(s)
  f2()

f1()


hr('Anonymus function lamda')
#Anonymus function

def cube(x): return x*x*x

#lambda

cube_v2 = lambda x: x*x*x

print(cube(3))
print(cube_v2(3))


hr('Recursive Function')
#Recursive Function
def factorial(x):
  if x == 1:
    return 1
  else:
    return (x * factorial(x-1))

print(factorial(5))

hr('Pass by Reference and Pass key value')
#Pass by Reference and Pass key value
def Pass(x):
  x[0] = 20

PassList = [10, 11, 12, 13, 14, 15]
Pass(PassList)

def myPass(x):
  x = [20, 30, 40]

myPass(PassList)
print(PassList)
```
