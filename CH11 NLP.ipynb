```python
char = str(input())
n = [0, 0, 0, 0]

for i in char:
  if i in pth.thai_consonants:
    n[0] = n[0]+1
  if i in pth.thai_vowels:
    n[1] = n[1]+1
  if i in pth.thai_digits:
    n[2] = n[2]+1
  n[3] = n[3]+1

print("จำนวนทั้งหมด", n[0]+n[1]+n[2])
print("พยัญชนะ ", n[0])
print("สระ", n[1])
print("ตัวเลข", n[2])


# for i in char:
#   if i in pth.thai_consonants:
#     n[0] = n[0]+1
#   n[3] = n[3]+1

# n[3] = 0
# for i1 in char:
#   if i1 in pth.thai_vowels:
#     n[1] = n[1]+1
#   n[3] = n[3]+1

# n[3] = 0
# for i2 in char:
#   if i2 in pth.thai_digits:
#     n[2] = n[2]+1
#   n[3] = n[3]+1
```
