```python
import numpy as np
import random

O = []
X = []

win = [
    [1, 2, 3],
    [4, 5, 9],
    [7, 8, 1],
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9],
    [1, 5, 9],
    [3, 5, 7]
]

def checkWin(P):
    for w in win:
        if all(x - 1 in P for x in w):
            return True
    return False

def displayOX():
    OX = np.array([' '] * 9)
    OX[O] = ['O']
    OX[X] = ['X']
    print(OX.reshape([3, 3]))

def AI():
    validmove = list(set(range(9)) - set(O + X))
    V = [-100] * 9
    for m in validmove:
        tempX = X + [m]
        V[m], criticalmove = evaloX(O, tempX)
        if len(criticalmove) > 0:
            move = [i - 1 for i in criticalmove if i - 1 in validmove]
            return random.choice(move)
    maxV = max(V)
    imaxV = [i for i, j in enumerate(V) if j == maxV]
    return random.choice(imaxV)

def evaloX(O, X):
    SO, SX, criticalmove = calSoX(O, X)
    return 1 + SX - SO, criticalmove

def calSoX(O, X):
    SO = SX = 0
    criticalmove = []
    for w in win:
        o = [i - 1 in O for i in w]
        x = [i - 1 in X for i in w]
        if not any(x):
            nO = o.count(True)
            SO += nO
            if nO == 2:
                print('critical', w)
                criticalmove = w
        if not any(o):
            SX += x.count(True)
    return SO, SX, criticalmove

while True:
    move = int(input('Choose position [1-9]')) - 1
    while move in O + X or move > 8 or move < 0:
        move = int(input('Bad move: Choose position [1-9]')) - 1
    O.append(move)
    displayOX()
    if checkWin(O):
        print('O win')
        break
    if len(O) + len(X) == 9:
        print('Draw')
        break
    X.append(AI())
    displayOX()
    if checkWin(X):
        print('X win')
        break
    if len(O) + len(X) == 9:
        print('Draw')
        break
```

```python
import numpy as np #นำเข้าโมดูล numpy และเปลี่ยนชื่อเป็น np เพื่อให้สามารถเรียกใช้ฟังก์ชันและอ็อบเจกต์ใน numpy ได้โดยการใช้ np แทนชื่อ numpy ในที่อื่นของโค้ด
import random # นำเข้าโมดูล random เพื่อสร้างตัวเลขสุ่มในการเล่นเกม.

win = [       #เป็นรายการของรายการ 3 ตัวเลขที่แทนตำแหน่งที่ชนะในเกม Tic-Tac-Toe ซึ่งเก็บรายการทุกที่เป็นไปได้ของการชนะ เช่น 3 ตำแหน่งแนวนอน 3 ตำแหน่งแนวตั้ง และ 2 ตำแหน่งแนวเฉียง
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9],
    [1, 5, 9],
    [3, 5, 7]
     ]

def checkWin(P):#เป็นฟังก์ชันที่ใช้ตรวจสอบว่าผู้เล่นที่กำลังทำทราบโดย P ชนะหรือไม่ โดยตรวจสอบว่ามีตำแหน่งที่ชนะในรายการ 'win' หรือไม่
    for w in win:
        if all(x - 1 in P for x in w):
            return True
    return False

def displayOX(O, X):#เป็นฟังก์ชันที่ใช้ในการแสดงสถานะปัจจุบันของเกม Tic-Tac-Toe โดย O แทนตำแหน่งที่ผู้เล่น O ได้เลือกและ X แทนตำแหน่งที่ผู้เล่น X ได้เลือก
    OX = np.array([' '] * 9)
    OX[O] = ['O']
    OX[X] = ['X']
    print(OX.reshape([3, 3]))

def AI(O, X):#เป็นฟังก์ชัน AI ที่ใช้ในการคำนวณการเลือกตำแหน่งของ AI ที่เป็นตัวเลือกต่อไป โดยอิงจากตำแหน่งที่ผู้เล่น O และ X เลือกไปแล้ว
    validmove = list(set(range(9)) - set(O + X))
    V = [-100] * 9
    for m in validmove:
        tempX = X + [m]
        V[m], criticalmove = evaloX(O, tempX)
        if len(criticalmove) > 0:
            move = [i - 1 for i in criticalmove if i - 1 in validmove]
            return random.choice(move)
    maxV = max(V)
    imaxV = [i for i, j in enumerate(V) if j == maxV]
    return random.choice(imaxV)

def evaloX(O, X):#เป็นฟังก์ชันที่ใช้ในการประเมินความเสี่ยงของการเลือกตำแหน่งที่มีโอกาสชนะสูงสุดสำหรับ AI โดยอิงจากตำแหน่งที่ผู้เล่น O และ X เลือกไปแล้ว
    SO, SX, criticalmove = calSOX(O, X)
    return 1 + SX - SO, criticalmove

def calSOX(O, X):# เป็นฟังก์ชันที่ใช้ในการคำนวณคะแนน SO (Score of O) และ SX (Score of X) ของผู้เล่น O และ X ตามรายการที่ชนะ win
    SO = SX = 0
    criticalmove = []
    for w in win:
        o = [i - 1 in O for i in w]
        x = [i - 1 in X for i in w]
        if not any(x):
            nO = o.count(True)
            if nO == 2:
                print('critical', w)
                criticalmove = w

        if not any(o):
            SX += x.count(True)
    return SO, SX, criticalmove

O = []#รายการ O เก็บตำแหน่งที่ผู้เล่น O ได้เลือก.
X = []#รายการ X เก็บตำแหน่งที่ผู้เล่น X ได้เลือก.

while True:#วนลูปไม่จบ.
    move = int(input('Choose position [1-9]')) - 1#รับตำแหน่งที่ผู้เล่นต้องการเลือก
    while move in O + X or move > 8 or move < 0:#ตรวจสอบว่าตำแหน่งที่ผู้เล่นเลือกถูกต้องหรือไม่
        move = int(input('Bad move: Choose position [1-9]')) - 1
    O.append(move)#เพิ่มตำแหน่งที่ผู้เล่น O เลือกลงในรายการ O
    displayOX(O, X)#แสดงสถานะปัจจุบันของเกม
    if checkWin(O):#ตรวจสอบว่าผู้เล่น O ชนะหรือไม่
        print('O win')# แสดงผลว่าผู้เล่น X ชนะ
        break
    if len(O) + len(X) == 9:#ตรวจสอบว่ามีการเล่นเต็มแล้วหรือไม่
        print('Draw')#แสดงผลว่าเกมเสมอ
        break

    X.append(AI(O, X))#เพิ่มตำแหน่งที่ AI เลือกลงในรายการ X
    displayOX(O, X)
    if checkWin(X):#ตรวจสอบว่าผู้เล่น X ชนะหรือไม่
        print('X win')# แสดงผลว่าผู้เล่น X ชนะ
        break
    if len(O) + len(X) == 9:#ตรวจสอบว่ามีการเล่นเต็มแล้วหรือไม่
        print('Draw')#แสดงผลว่าเกมเสมอ
        break
```
