```python
import numpy as np
data = [[2, 6],
        [4, 10],
        [6, 11],
        [10, 22],
        [12, 25]]
# print("data =", data)
data = np.array(data)
# print("data =", data)
# x2 = data[:, 0]
# print("data =", x2)
x = data[:, 0].reshape(-1, 1)
# print("data =", x)
y = data[:, 1].reshape(-1, 1)
print("data =", y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)

y_5 = model.predict([[5]])
print("y เมื่อ x = 5 มีค่า", y_5)
y_9 = model.predict([[9]])
print("x เมื่อ y = 9 มีค่า", y_9)
```

```python
print("ค่า coefficient หรือค่า a = ", model.coef_[0])
print("ค่า intercept หรือค่า b = ", model.intercept_)
```

```python
from sklearn.metrics import r2_score

y_predicted = model.predict(x)

score = r2_score(y, y_predicted)

print("y_predicted = ", y_predicted)
print("R-Squared = ", score)
accuracy = "{:.2f}".format(score * 100)
print(f"ค่าความแม่นยำคือ {accuracy}%")
```
